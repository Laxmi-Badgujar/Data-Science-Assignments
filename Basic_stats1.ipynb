## Interpretation-

2. Also the representation helps understand that the data is not normally distributed as significant diiference is seen in median (50%) and 75% which also indicates that the data is right skewed.
3. The variables Avg Price, Total Sales Value and Net sales Value have more scattered data and consists of outliers.
4.  The variables Volume and Discount Rate (%) have interquartile range the minimum stating that the data is centered around the mean
